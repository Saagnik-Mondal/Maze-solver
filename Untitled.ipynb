```python
import subprocess
import json
import time
```

```python
def create_maze(width, height):
    maze = np.ones((height * 2 + 1, width * 2 + 1), dtype=int)
    
    def carve_path(x, y):
        maze[y * 2 + 1, x * 2 + 1] = 0
        
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        np.random.shuffle(directions)
        
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            
            if 0 <= nx < width and 0 <= ny < height and maze[ny * 2 + 1, nx * 2 + 1] == 1:
                maze[y * 2 + 1 + dy, x * 2 + 1 + dx] = 0
                carve_path(nx, ny)
    
    carve_path(0, 0)
    return maze
```

```python
def solve_maze(maze, start, end):
    solution = np.zeros_like(maze)
    visited = np.zeros_like(maze, dtype=bool)
    
    def backtrack(pos):
        y, x = pos
        
        if pos == end:
            solution[y, x] = 1
            return True
        
        if visited[y, x] or maze[y, x] == 1:
            return False
        
        visited[y, x] = True
        solution[y, x] = 1
        
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        
        for dy, dx in moves:
            ny, nx = y + dy, x + dx
            
            if 0 <= ny < maze.shape[0] and 0 <= nx < maze.shape[1]:
                if backtrack((ny, nx)):
                    return True
        
        solution[y, x] = 0
        return False
    
    backtrack(start)
    return solution
```

```python
def run_pygame_solver(maze, start, end, cell_size=20):
    """Run pygame solver in separate Python process"""
    
    # Prepare data to pass to the pygame script
    data = {
        'maze': maze.tolist(),
        'start': list(start),
        'end': list(end),
        'cell_size': cell_size
    }
    
    # Convert to JSON string
    json_data = json.dumps(data)
    
    print("Launching pygame visualization...")
    print(f"Maze size: {maze.shape}")
    print(f"Start: {start}, End: {end}")
    print("\nThe pygame window will open shortly.")
    print("Close the window when you're done viewing the solution.\n")
    
    # Run the pygame script in a separate process
    try:
        result = subprocess.run(
            [sys.executable, 'maze_solver_pygame.py', json_data],
            capture_output=True,
            text=True,
            cwd='/Volumes/X9 Pro/Resume Project/Forest Disection'
        )
        
        # Print the output
        print(result.stdout)
        
        if result.stderr:
            print("Errors:", result.stderr)
        
        # Parse results if available
        for line in result.stdout.split('\n'):
            if line.startswith('RESULTS_JSON:'):
                results = json.loads(line.replace('RESULTS_JSON:', '').strip())
                return results
        
        return None
        
    except Exception as e:
        print(f"Error running pygame: {e}")
        return None
```

```python
def visualize_maze(maze, solution=None, start=None, end=None):
    display = maze.copy().astype(float)
    
    if solution is not None:
        display = display + solution * 2
    
    if start is not None:
        display[start] = 3
    
    if end is not None:
        display[end] = 4
    
    colors = ['white', 'black', 'lightblue', 'green', 'red']
    cmap = ListedColormap(colors)
    
    plt.figure(figsize=(10, 10))
    plt.imshow(display, cmap=cmap, interpolation='nearest')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
maze = create_maze(15, 15)
print(f"Maze shape: {maze.shape}")

visualize_maze(maze)
```

```python
start_pos = (1, 1)
end_pos = (maze.shape[0] - 2, maze.shape[1] - 2)

solution = solve_maze(maze, start_pos, end_pos)

visualize_maze(maze, solution, start_pos, end_pos)
```
